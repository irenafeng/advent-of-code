```python
input_file = open("input03.txt", "r")
input = input_file.read().split("\n")
input.pop()

sample = ["467..114..", "...*......", "..35..633.", "......#...", "617*......", ".....+.58.", "..592.....", "......755.", "...$.*....", ".664.598.."]
```

```python
numberlist = ["1", "2", "3", "4", "5", "6", "7", "8", "9", "0"]

input_matrix = []
for line in input:
    letters = [x for x in line]
    input_matrix.append(letters)

sample_matrix = []
for line in sample:
    letters = [x for x in line]
    sample_matrix.append(letters)
```

```python
# PART ONE

width = len(input[0])
height = len(input)

symbols = [[False] * width for _ in range(height)]
numbers = []

def settrue(y, x):
    if 0 <= x < width and 0 <= y < height:
        symbols[y][x] = True

for i, line in enumerate(input):
    for j, char in enumerate(line):
        if (char not in numberlist) and char != ".":
            settrue(i-1, j-1)
            settrue(i-1, j)
            settrue(i-1, j+1)
            settrue(i+1, j-1)
            settrue(i+1, j)
            settrue(i+1, j+1)
            settrue(i, j-1)
            settrue(i, j+1)
    
    last = "."
    start = -1
    for j, char in enumerate(line):
        if char in numberlist:
            if last == ".":
                start = j
                last = char
            else:
                last += char    
        else:
            if last != ".":
                numbers.append((i, start, last))
                last = "."
    if last != ".":
        numbers.append((i, start, last))
        last = "."

print(numbers)

parts_sum = 0
for num in numbers:
    adj = False
    for numlength in range(len(num[2])):
        if symbols[num[0]][num[1]+numlength]:
            adj = True
    if adj:
        parts_sum += int(num[2])

print("total: ", parts_sum)
```

```python
# PART TWO

parts = [[-1] * width for _ in range(height)]
numbers_2 = []

print(numberlist)

id = 0
for i, line in enumerate(input):
    last = "."
    start = -1

    for j, char in enumerate(line):
        if char in numberlist:
            if last == ".":
                start = j
                last = char
            else:
                last += char
            parts[i][j] = id
        else:
            if char == "*":
                parts[i][j] = -2
            if last != ".":
                numbers_2.append(last)
                last = "."
                id += 1
    if last != ".":
        numbers_2.append(last)
        last = "."
        id += 1

def getpart(y, x):
    if 0 <= x < width and 0 <= y < height:
        return parts[y][x]
    return -1

total = 0
for i, line in enumerate(parts):
    for j, part in enumerate(line):
        #print(part)
        if part == -2:
            adjset = {
                getpart(i-1, j-1),
                getpart(i-1, j+1),
                getpart(i+1, j-1),
                getpart(i+1, j),
                getpart(i+1, j+1),
                getpart(i, j-1),
                getpart(i-1, j),
                getpart(i, j+1)
            }
            adj = [a for a in adjset if a >= 0]
            #print(adj)
            if len(adj) == 2:
                total += int(numbers_2[adj[0]]) * int(numbers_2[adj[1]])

print(parts)
print(numbers_2)

print("total: ", total)
```

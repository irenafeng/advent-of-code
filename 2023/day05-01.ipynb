```python
# PART ONE

# seeds_list = [int(i) for i in sample[0].split(": ")[1].split()]

# seed_soil_map = sample.index("seed-to-soil map:")
# soil_fertilizer_map = sample.index("soil-to-fertilizer map:")
# fertilizer_water_map = sample.index("fertilizer-to-water map:")
# water_light_map = sample.index("water-to-light map:")
# light_temp_map = sample.index("light-to-temperature map:")
# temp_humid_map = sample.index("temperature-to-humidity map:")
# humid_location_map = sample.index("humidity-to-location map:")

# seed_soil = sample[seed_soil_map+1:soil_fertilizer_map-1]
# soil_fert = sample[soil_fertilizer_map+1:fertilizer_water_map-1]
# fert_water = sample[fertilizer_water_map+1:water_light_map-1]
# water_light = sample[water_light_map+1:light_temp_map-1]
# light_temp = sample[light_temp_map+1:temp_humid_map-1]
# temp_humid = sample[temp_humid_map+1:humid_location_map-1]
# humid_location = sample[humid_location_map+1:len(sample)]

seeds_list = [int(i) for i in input[0].split(": ")[1].split()]

seed_soil_map = input.index("seed-to-soil map:")
soil_fertilizer_map = input.index("soil-to-fertilizer map:")
fertilizer_water_map = input.index("fertilizer-to-water map:")
water_light_map = input.index("water-to-light map:")
light_temp_map = input.index("light-to-temperature map:")
temp_humid_map = input.index("temperature-to-humidity map:")
humid_location_map = input.index("humidity-to-location map:")

seed_soil = input[seed_soil_map+1:soil_fertilizer_map-1]
soil_fert = input[soil_fertilizer_map+1:fertilizer_water_map-1]
fert_water = input[fertilizer_water_map+1:water_light_map-1]
water_light = input[water_light_map+1:light_temp_map-1]
light_temp = input[light_temp_map+1:temp_humid_map-1]
temp_humid = input[temp_humid_map+1:humid_location_map-1]
humid_location = input[humid_location_map+1:len(input)]

def conversions(list, map):
    source_starts = []
    source_ends = []

    for line in map:
        source_starts.append(int(line.split()[1]))
        source_ends.append(int(line.split()[1]) + int(line.split()[2]) - 1)
    
    new_list = [int() for _ in range(len(list))]

    for i, list_item in enumerate(list):
        source = -1
        index = 0
        for j, start in enumerate(source_starts):
            if list_item >= start and list_item <= source_ends[j]:
                source = start
                index = j
        
        if source == -1:
            new_list[i] = list[i]
        else:
            new_list[i] = list[i] - source + int(map[index].split()[0])
    
    return(new_list)

soil_list = conversions(seeds_list, seed_soil)
fert_list = conversions(soil_list, soil_fert)
water_list = conversions(fert_list, fert_water)
light_list = conversions(water_list, water_light)
temp_list = conversions(light_list, light_temp)
humid_list = conversions(temp_list, temp_humid)
locations = conversions(humid_list, humid_location)

print("seeds:     ", seeds_list)
print("soil:      ", soil_list)
print("fert:      ", fert_list)
print("water:     ", water_list)
print("light:     ", light_list)
print("temp:      ", temp_list)
print("humid:     ", humid_list)
print("locations: ", locations)

print("")
print("lowest location: ", min(locations))
```
